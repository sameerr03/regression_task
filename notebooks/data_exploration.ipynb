### Preliminary data analysis
To understand what could make a good regression model to predict fuel consumption, we want to look at the various features that are provided in the training dataset. We can understand the relationship between the features and the target variable by using a correlation matrix and by creating scatterplots between the features and the targets.

We use the features ENGINE SIZE, CYLINDERS, COEMISSIONS and FUEL CONSUMPTION as these are in numerical form and will not require the creation of dummy variables

```python
# Select the features we're interested in
features = ['ENGINE SIZE', 'CYLINDERS', 'FUEL CONSUMPTION', 'COEMISSIONS ']
df_selected = df[features]
```

```python
# Create a correlation matrix
correlation_matrix = df_selected.corr()

correlation_matrix
```

```python
# Create scatter plots
fig, axes = plt.subplots(3, 1, figsize=(10, 15))
fig.suptitle('Feature Relationships with Fuel Consumption', fontsize=16)

features_to_plot = ['ENGINE SIZE', 'CYLINDERS', 'COEMISSIONS ']
for i, feature in enumerate(features_to_plot):
    axes[i].scatter(df[feature], df['FUEL CONSUMPTION'])
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('FUEL CONSUMPTION')
    axes[i].set_title(f'{feature} vs FUEL CONSUMPTION')

plt.tight_layout()
```

### Analysis of Categorical variables
We need to analyse the categorical variables in order to indentify if there are potentially useful features among them. 

```python
# Print distribution of categorical variables
categorical_vars = ['VEHICLE CLASS', 'TRANSMISSION', 'FUEL']

print("\nDistribution of Categorical Variables:")
for var in categorical_vars:
    print(f"\n{var}:")
    value_counts = df[var].value_counts()
    total_count = len(df)
    for category, count in value_counts.items():
        percentage = (count / total_count) * 100
        print(f"{category}: {count} ({percentage:.2f}%)")
```

```python
fig, axes = plt.subplots(len(categorical_vars), 1, figsize=(12, 6*len(categorical_vars)))
fig.suptitle('Fuel Consumption by Categorical Variables', fontsize=16)

for i, var in enumerate(categorical_vars):
    sns.boxplot(x=var, y='FUEL CONSUMPTION', data=df, ax=axes[i])
    axes[i].set_title(f'Fuel Consumption by {var}')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
```

### Conclusion
From the numerical variables, it seems that all of them are highly correlated with each other and the target variable. Choosing only one out of these variables therefore makes sense as otherwise our regression will have multicolinearity issues, and we will violate the assumption of feature independence. In terms of the categorical variables, the boxplots of transmission type indicate that there probably is no significant explanatory power gained from adding this variable to our model. The boxplots of Fuel type show promise, however types D,N,E have 8 observations collectively, which is far too less and we risk overfitting the model to the training data if we use this. Vehicle class appears promising to be used in the regression model. We could employ binary variables to indicate the class. We could also try combining vehicle classes that appear to have similar fueal consumption, like compact and subcompact or van-cargo and van-passenger.
